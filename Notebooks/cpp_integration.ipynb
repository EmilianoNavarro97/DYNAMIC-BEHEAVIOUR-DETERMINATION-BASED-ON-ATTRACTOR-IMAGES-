```python
import struct
import matplotlib.pyplot as plt
from shutil import copy
```

```python
with open('/home/nubol23/Documents/ExportChaos/0_134.bin', 'rb') as f:
    data = struct.unpack('d'*(3*240000), f.read())

elevation, azimuth = 45, 45

ax = plt.axes(projection='3d')
ax.plot3D(data[::3], data[1::3], data[2::3], 'black', linewidth=0.5)
ax.set_axis_off()
ax.view_init(elev=elevation, azim=azimuth)
    
plt.show()
```
