```python
import matplotlib.pyplot as plt
import numpy as np
import os
from tqdm import tqdm
```

```python
%matplotlib notebook
```

```python
n = 1200
for col in tqdm(range(n)):
    x = np.load(f'../X_axis/{col}.npy')
    points1 = []
#     points2 = []
    for i in range(1, len(x)-1):
        if x[i-1] < x[i] and x[i] > x[i+1]:
            points1.append(x[i])
            
#         if x[i-1] > x[i] and x[i] < x[i+1]:
#             points2.append(x[i])
    plt.plot([col]*len(points1), points1, 'ro', markersize=0.1)
```
